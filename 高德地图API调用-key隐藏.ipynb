```python
import requests
import pandas as pd
import time
```

```python
key=''#填写自己的key
def geo(address:str,city=None)-> dict:
    parameters={'key':key,
                'city':city,
                'citylimit':True,
                'address':address}
    r=requests.get("https://restapi.amap.com/v3/geocode/geo?parameters",params=parameters)
    time.sleep(2)#防止key被锁
    data=r.json()['geocodes'][0]
    df1=pd.DataFrame.from_dict(data,orient='index').T
    return df1
```

```python
def processText(flname):
    data=pd.read_excel(flname,encoding='utf-8')
    address=data['居住地'].drop_duplicates()
    df=pd.DataFrame()
    for i in address:
        df_i=geo(i)
        df_i['居住地']=i
        df=pd.concat([df,df_i])
    a=data.merge(df,on=['居住地'])
    return a
```

```python
df=processText('2022424上海疫情爬虫数据.xlsx')
```
